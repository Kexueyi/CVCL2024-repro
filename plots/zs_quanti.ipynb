```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

result_folder_1 = "clip_awa2_07-30_22-34"
df_1 = pd.read_csv(f'/home/xke001/demo/zero-shot/results/{result_folder_1}/per_class_accuracy.csv')
result_folder_2 = "clip_awa2_07-30_22-25"
df_2 = pd.read_csv(f'/home/xke001/demo/zero-shot/results/{result_folder_2}/per_class_accuracy.csv')
```

```python
# Set up the plotting area
fig, ax = plt.subplots(figsize=(12, 8))

# 绘制折线图
ax.plot(df_1['Clean_Class_Name'], df_1['Accuracy'], label='Without Prefix', marker='o', linestyle='-')
ax.plot(df_2['Clean_Class_Name'], df_2['Accuracy'], label='With Prefix', marker='x', linestyle='-')

# 添加标签和标题
ax.set_xlabel('Class Names')
ax.set_ylabel('Accuracy')
ax.set_title('Comparison of Class Accuracy with and without Prefix')
ax.legend()

# 改善布局以容纳所有类名
plt.xticks(rotation=90)

plt.show()
```
